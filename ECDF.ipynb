<h1> ECDF analysis of different distributions </h1>

```python
import analysis
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

<h2> Configuration </h2>

```python
size = 1000  # Set sample size
```

<h2> ECDF Random Sample </h2>

```python
data = np.random.random_sample(size=size)
```

```python
x, y = analysis.ecdf(data)
```

```python
plt.plot(x,y)
```

```python
plt.show()
```

<h2> ECDF Random Normal Distribution </h2>

```python
normal = np.random.normal(size=size)
x, y = analysis.ecdf(normal)
plt.plot(x,y)
plt.show()
```

<h2> ECDF Random Exponential Distribution </h2>

```python
exp = np.random.exponential(size=size)
x, y = analysis.ecdf(exp)
plt.plot(x,y)
plt.show()
```

<h2> ECDF Random Lognormal Distribution </h2>

```python
logn = np.random.lognormal(size=size)
x, y = analysis.ecdf(logn)
plt.plot(x,y)
plt.show()
```
